```python
import tkinter as tk
from datetime import datetime
import pygame
```

```python
def set_alarm():
    alarm_time = alarm_entry.get()
    alarm_label.config(text=f"Alarm set for {alarm_time}", fg="green", font=("Arial", 10, "italic"))

def check_alarm():
    current_time = datetime.now().strftime("%H:%M")
    current_time_label.config(text=f"Current Time: {current_time}", fg="blue", font=("Arial", 12))
    
    alarm_time = alarm_entry.get()
    if current_time == alarm_time:
        alarm_label.config(text="Alarm Ringing!", fg="red", font=("Arial", 14))
        try:
            pygame.mixer.init()
            pygame.mixer.music.load(r"C:\Users\S.A Computer\Documents\Alarm clock\mixkit-classic-alarm-995.wav")
            pygame.mixer.music.play()
            while pygame.mixer.music.get_busy():
                continue
        except Exception as e:
            print(f"Sound could not be played: {e}")

    root.after(1000, check_alarm)

root = tk.Tk()
root.title("Alarm Clock")

frame = tk.Frame(root, padx=20, pady=20)
frame.pack()

label = tk.Label(frame, text="Enter Alarm Time (HH:MM):", font=("Arial", 12))
label.grid(row=0, column=0)

alarm_entry = tk.Entry(frame, font=("Arial", 12))
alarm_entry.grid(row=0, column=1)

set_button = tk.Button(frame, text="Set Alarm", command=set_alarm, font=("Arial", 10, "bold"))
set_button.grid(row=1, columnspan=2, pady=10)

current_time_label = tk.Label(frame, text="Current Time: ", font=("Arial", 12))
current_time_label.grid(row=2, columnspan=2)

alarm_label = tk.Label(frame, text="", font=("Arial", 14))
alarm_label.grid(row=3, columnspan=2)

root.after(1000, check_alarm)
root.mainloop()
```
